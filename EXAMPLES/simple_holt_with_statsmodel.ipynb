```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
from matplotlib import pyplot as plot

prices = [100, 100, 120, 112, 100, 105, 98, 102, 110, 108, 107, 111, 109, 112, 115, 113]

model = ExponentialSmoothing(prices, trend='add', seasonal=None)
model_mul = ExponentialSmoothing(prices, trend='mul', seasonal=None)
fit_model = model.fit(smoothing_level=0.3, smoothing_trend=0.1)
fit_model_mul = model_mul.fit(smoothing_level=0.3, smoothing_trend=0.1)
smoothed = fit_model.fittedvalues
smoothed_mul = fit_model_mul.fittedvalues

figure = plot.figure(figsize=(10, 6))

forcast = fit_model.forecast(5)
forcast_mul = fit_model_mul.forecast(5)


plot.plot(prices, label='Dane Oryginalne', marker='o', color='blue')
plot.plot(smoothed, label='Wygładzone (Addytywne)', color='orange')
plot.plot(smoothed_mul, label='Wygładzone (Mnożnikowe)', color='green')
plot.plot(range(len(prices), len(prices) + 5), forcast, label='Prognoza (Addytywne)', marker='o', linestyle='--', color='orange')
plot.plot(range(len(prices), len(prices) + 5), forcast_mul, label='Prognoza (Mnożnikowe)', marker='o', linestyle='--', color='green')
plot.title('Wygładzanie wykładnicze Holta-Wintersa za pomocą statsmodels')
plot.xlabel('Okresy')
plot.ylabel('Wartości')
plot.grid(True)
plot.legend()
plot.show()
```
